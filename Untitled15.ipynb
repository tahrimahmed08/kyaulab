```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def dda(x0, y0, x1, y1):
    dx = x1 - x0
    dy = y1 - y0
    step = max(abs(dx), abs(dy))
    x_inc = dx / step
    y_inc = dy / step
    x = x0
    y = y0
    dict = {'x': [x], 'y': [y], 'round(x)': [x], 'round(y)': [y]}
    for i in range(step):
        yield round(x), round(y)
        x += x_inc
        y += y_inc
        dict['x'].append(x)
        dict['y'].append(y)
        dict['round(x)'].append(round(x))
        dict['round(y)'].append(round(y))
    df = pd.DataFrame(dict)
    print(df)

def main():
    x0 = 10
    y0 = 10
    x1 = 17
    y1 = 16
    x, y = zip(*dda(x0, y0, x1, y1))
    plt.title('DDA algorithm')
    plt.plot(x, y)
    plt.show()
main()
```

```python
def bressenham(x0, y0, x1, y1):
    x = x0
    y = y0
    dx = x1 - x0
    dy = y1 - y0
    p = 2 * dy - dx
    dict = {'p': [p], 'x': [x], 'y': [y]}
    while True:
        yield x, y
        if x == x1 and y == y1:
            break
        if p > 0:
            x += 1
            y += 1
            p += 2 * dy - 2 * dx
        else:
            x += 1
            p += 2 * dy
        dict['p'].append(p)
        dict['x'].append(x)
        dict['y'].append(y)
    df = pd.DataFrame(dict)
    print(df)

def main():
    x0 = 10
    y0 = 10
    x1 = 17
    y1 = 16
    x, y = zip(*bressenham(x0, y0, x1, y1))
    plt.title('Bressenham algorithm')
    plt.plot(x, y)
    plt.show()

main()
```

```python
def draw_circle(radius):
    x = 0
    y = radius
    d = 3 - 2 * radius
    points = set()

    while x <= y:
        points.add((x, y))
        points.add((x, -y))
        points.add((-x, y))
        points.add((-x, -y))
        points.add((y, x))
        points.add((y, -x))
        points.add((-y, x))
        points.add((-y, -x))

        if d < 0:
            d += 4 * x + 6
        else:
            d += 4 * (x - y) + 10
            y -= 1
        x += 1

    for point in points:
        plt.scatter(*point, color='black')

    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Bressenham Circle algorithm')
    plt.show()

radius = 10
draw_circle(radius)
```

```python
def draw_circle(radius):
    x = 0
    y = radius
    d = 1 - radius

    points = set()

    while x <= y:
        points.update([(x, y), (x, -y), (-x, y), (-x, -y),
                       (y, x), (y, -x), (-y, x), (-y, -x)])

        if d < 0:
            d += 2 * x + 3
        else:
            d += 2 * (x - y) + 5
            y -= 1
        x += 1

    plt.scatter(*zip(*points), color='black')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Midpoint Circle drawing algorithm')
    plt.show()
radius = 5
draw_circle(radius)
```
